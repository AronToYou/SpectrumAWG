```python
import matplotlib.pyplot as plt
from math import log10
```

```python
att = 20 # Attenuation of Signal (dBm)

### Raw Data ###
## Trap Number Sweeps ##
# Number of traps #
N = [i for i in range(1, 31)]
# Peak Power (dBm); Max over AOM Bandwidth #
P = [13, 7, 4, 1, -1, -2, -3.7, -4, -8, -9, -9, -9.5, -10, -11, -12, -12.5, -13,
     -13.5, -14, -14, -14, -14.5, -14.8, -15, -15.8, -16, -16.3, -16.8, -17.3, -17.5]

## Voltage Sweeps ##
# Programmed Output Voltage (mV) #
mV = [i*100 for i in range(1, 11)] 
# P_% is the power for % traps (dBm) #
P_30 = [-27, -22, -20, -17.5, -16, -14, -13, -11, -10.5, -10]
P_20 = [-24, -18, -16, -14, -12, -10.5, -9, -8.5, -7, -5]
P_10 = [-18, -14, -10, -8, -7, -4, -3, -2, -1, -0.3]
P_5 = [-14, -8.3, -3, -1.3, 0, 2, 3]
P_3 = [-10, -4, 1, 3, 5]
P_1 = [0.5, 7, 10, 12]
Ps = [P_30, P_20, P_10, P_5, P_3, P_1]
Ns = [30, 20, 10, 5, 3, 1]
```

```python
plt.plot(N, P) # Peak values taken from AOM Bandwidth
plt.plot(N, [p + 10*log10(n) for n, p in zip(N, P)]) # Peak values taken from AOM Bandwidth
plt.title("Peak Power vs. # of Traps\nOutput Voltage = 400mV")
plt.xlabel("# of Traps")
plt.ylabel("Peak Power (dBm)")
plt.legend(["Per Trap", "Total Signal"])
plt.show()



plt.plot(N, [(10**(p/10)) for n, p in zip(N, P)])
plt.plot(N, [n*(10**(p/10)) for n, p in zip(N, P)])
plt.title("Peak Power vs. # of Traps\nOutput Voltage = 400mV")
plt.xlabel("# of Traps")
plt.ylabel("Peak Power (mW)")
plt.legend(["Per Trap", "Total Signal"])
plt.show()
```

```python
for Pi in Ps:
    plt.plot(mV[0:len(Pi)], Pi)
plt.title("Peak Power vs. Programmed Vout\nPer Trap")
plt.xlabel("Programmed Vout (mV)")
plt.ylabel("Peak Power (dBm)")
plt.ylim((-28, 12))
plt.legend(["30", "20", "10", "5", "3", "1"], title="# of Traps")
plt.show()

for n, Pi in zip(Ns, Ps):
    ofs = 10*log10(n)
    plt.plot(mV[0:len(Pi)], [p + ofs for p in Pi])
plt.title("Peak Power vs. Programmed Vout\nPer Trap * (# of Traps)")
plt.xlabel("Programmed Vout (mV)")
plt.ylabel("Peak Power (dBm)")
plt.ylim((-28, 12))
plt.legend(["30", "20", "10", "5", "3", "1"], title="# of Traps")
plt.show()
```

```python
for Pi in Ps:
    mW = [10**(p/10) for p in Pi]
    plt.plot(mV[0:len(mW)], mW)
plt.title("Peak Power vs. Programmed Vout\nPer Trap")
plt.xlabel("Programmed Vout (mV)")
plt.ylabel("Peak Power (mW)")
plt.legend(["30", "20", "10", "5", "3", "1"], title="# of Traps")
plt.show()

for n, Pi in zip(Ns, Ps):
    mW = [n*10**(p/10) for p in Pi]
    plt.plot(mV[0:len(mW)], mW)
plt.title("Peak Power vs. Programmed Vout\nPer Trap*(# of Traps)")
plt.xlabel("Programmed Vout (mV)")
plt.ylabel("Peak Power (mW)")
plt.legend(["30", "20", "10", "5", "3", "1"], title="# of Traps")
plt.show()
```

```python
plt.plot(mV, [10**(p/10) for p in P_30])
plt.title("Peak Power vs. Programmed Vout\n30 Traps")
plt.xlabel("Programmed Vout (mV)")
plt.ylabel("Peak Power (mW)")
plt.show()

plt.plot(mV[0:len(P_1)], [10**(p/10) for p in P_1])
plt.title("Peak Power vs. Programmed Vout\n1 Trap")
plt.xlabel("Programmed Vout (mV)")
plt.ylabel("Peak Power (mW)")
plt.show()
```
